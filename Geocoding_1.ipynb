```python
import pandas as pd
import numpy as np
from geopy import geocoders
pd.options.mode.chained_assignment = None
```

```python
g = geocoders.GoogleV3(api_key=None)
#https://geopy.readthedocs.io/en/stable/
```

```python
#define function using google api to get coordinates
def coordinate(input_add):
    location = g.geocode(input_add)
    return location.latitude, location.longitude
```

```python
#testint with city of toronto dataset
#https://www.toronto.ca/city-government/data-research-maps/open-data/open-data-catalogue/#1b3c08c4-0316-fc59-88c6-e6d7178a2cf8
address=pd.read_csv('facility_address.csv')
```

```python
a1=address[40:50]
a1.head()
```

```python
a1['coordinate']=a1['Full Address'].apply(coordinate)
```

```python
a1[['FACILITY NAME','Full Address','coordinate']]
```

```python
#distance calculation function-set target then loop all dataframe to calculate distance to target
from geopy import distance
```

```python
def distance_c(target,sds):
    import geopy
    from geopy import distance
    return geopy.distance.vincenty(target,sds)
```

```python
target=a1[a1['FACILITY NAME']=='Scarlett Heights Entrepreneurial Academy'].coordinate
```

```python
a1['distance']=a1['coordinate'].apply(lambda x: distance_c(target,x))
```

```python
#all facilities within 5km from the target
a1[a1['distance']<=5]
```
