```python
import pandas as pd
import numpy as np
from geopy import geocoders
pd.options.mode.chained_assignment = None
```

```python
g = geocoders.GoogleV3(api_key=None)
#https://geopy.readthedocs.io/en/stable/
```

```python
#define function using google api to get coordinates
def coordinate(input_add):
    location = g.geocode(input_add)
    return location.latitude, location.longitude
```

```python
#testint with city of toronto dataset
#https://www.toronto.ca/city-government/data-research-maps/open-data/open-data-catalogue/#1b3c08c4-0316-fc59-88c6-e6d7178a2cf8
address=pd.read_csv('facility_address.csv')
```

```python
a1=address[20:30]
a1.head()
```

```python
a1['coordinate']=a1['Full Address'].apply(coordinate)
```
